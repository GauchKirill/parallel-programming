```python
import numpy as np
import matplotlib.pyplot as plt

def show_solution(filename="file.txt"):
  """
  Считывает матрицу из файла и отображает все десятые строки на одном графике
  в виде зависимости "U(x)".

  Args:
    filename: Имя файла с матрицей.
  """
  try:
    # Считываем матрицу из файла, используя numpy
    matrix = np.loadtxt(filename)

    # Определяем количество строк в матрице
    num_rows = matrix.shape[0]
    num_columns = matrix.shape[1]

    # Создаем общий график
    plt.figure(figsize=(10, 8))  # Задаем размер общего графика

    # Проходим по каждой десятой строке и добавляем на график
    idx = np.linspace(0, num_rows-1, 5, endpoint=True)
    idx = idx.astype(int)
    for i in idx:
      # Получаем текущую строку
      row = matrix[i]

      # Рисуем зависимость U(x), добавляем подпись к линии
      plt.plot(row, label=f"t = {(i+1)/num_rows:.2f}T")

    # Настраиваем общий график
    plt.xlabel("x")  # Подписываем ось X
    plt.ylabel("U(x)")  # Подписываем ось Y
    plt.title("U(x)")  # Задаем заголовок графика
    plt.grid(True)  # Добавляем сетку
    plt.legend()  # Отображаем легенду
    plt.tight_layout()  # Оптимизируем расположение элементов графика

    # Показываем общий график
    plt.show()

  except FileNotFoundError:
    print(f"Ошибка: Файл '{filename}' не найден.")
  except Exception as e:
    print(f"Ошибка при обработке файла: {e}")

# Пример использования
if __name__ == "__main__":
  show_solution("file.txt")
```

```python
name = "solution.txt"
show_solution(name)
```
